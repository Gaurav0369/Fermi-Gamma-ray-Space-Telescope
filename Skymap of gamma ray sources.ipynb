```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.io import ascii
import astropy.coordinates as coord
import astropy.units as u
```

```python
# Load the dataset into a pandas DataFrame
GS = ascii.read("4FGL_DR3_Dataset_Enigma.csv",data_start = 2)
GS
```

```python
# Extract the coordinates of the sources
ra = coord.Angle(GS['col3'],unit=u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coord.Angle(GS['col4'],unit=u.degree)
```

```python
#Create a scatter plot of the sources on a skymap
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
```
